```python
import pandas as pd
import plotly.express as px
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
housing_price_filepath = "/content/California_Houses.csv"
```

```python
housingprice_data = pd.read_csv(housing_price_filepath)
```

```python
housingprice_data.head()
```

```python
plt.figure(figsize= (14,6))

sns.lineplot(data = housingprice_data['Tot_Rooms'] ,label = 'Total Rooms')

sns.lineplot(data = housingprice_data['Tot_Bedrooms'],label = 'Total Bedrooms')
```

```python
plt.figure(figsize=(10,6))
cond = housingprice_data.query('Median_Income <6').Median_Income
sns.barplot(x= housingprice_data.Median_Age, y=cond)
```

```python
plt.figure(figsize=(14,7))
sns.heatmap(data = housingprice_data.corr(), annot =True)
```

```python
plt.figure(figsize=(14,7))
sns.scatterplot(x=housingprice_data['Median_Income'], y=housingprice_data['Median_House_Value'], hue=housingprice_data['Median_Age'])
```

```python
plt.figure(figsize=(14,7))
cond_age = housingprice_data.query('30 <= Median_Age <= 40').Median_Age
sns.kdeplot(data=housingprice_data, x='Distance_to_LA', hue=cond_age, shade=True)
```

```python
fig = px.scatter(housingprice_data, x="Median_House_Value", y="Median_Income",color ='Households')
fig.show()
```
