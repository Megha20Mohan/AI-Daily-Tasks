```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
house_data=pd.read_csv('California_Houses.csv')
```

```python
house_data.head()
```

# TASK-1

```python
sns.lineplot(data=house_data['Tot_Rooms'], label="Total Rooms")


sns.lineplot(data=house_data['Tot_Bedrooms'], label="Total Bedrooms")
```

# TASK-2

```python
sns.barplot(x=house_data.Median_Age, y=house_data.query('Median_Income<6').Median_Income)

#when u need to put conditions in we use query method and in it put the condition and the column
```

# TASK-3

```python
a=house_data.corr()
sns.heatmap(a,annot=True)
```

# TASK-4

```python
sns.scatterplot(x=house_data['Median_Income'], y=house_data['Median_House_Value'], hue=house_data['Median_Age'])
```

# TASK-5

```python
sns.kdeplot(data=house_data, x=house_data.Distance_to_LA, hue=house_data.query('30<=Median_Age<=40').Median_Age, shade=True)
```
