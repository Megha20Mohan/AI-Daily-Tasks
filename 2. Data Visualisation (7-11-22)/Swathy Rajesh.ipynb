```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('C:\\Users\\swath\\Documents\\AI CLUB\\California_Houses.csv')
display(df)
```

```python
sns.lineplot(data=df['Tot_Rooms'])
sns.lineplot(data=df['Tot_Bedrooms'])
```

```python
plt.figure(figsize=(10,6))
v=df.query('Median_Income <6').Median_Income
sns.barplot(x=df['Median_Age'],y=v)
```

```python
plt.figure(figsize=(14,7))
plt.title("Heat Map")
matrix = df.corr()
sns.heatmap(data= matrix,annot=True)
```

```python
plt.figure(figsize=(14,7))
sns.scatterplot(x=df['Median_Income'], y=df['Median_House_Value'], hue=df['Median_Age'])
```

```python
plt.figure(figsize=(14,7))
cond_age = df.query('30 <= Median_Age <= 40').Median_Age
sns.kdeplot(data=df, x='Distance_to_LA', hue=cond_age, shade=True)
```

```python
import plotly.express as pl
plt.figure(figsize=(8,6))
fig = pl.scatter(df, x='Households',y='Population', color="Median_Income")
fig.show()
```
