```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('California_Houses.csv')
```

```python
df.head()
```

```python
#1
sns.lineplot(data=df['Tot_Rooms'], label="Total Rooms")
sns.lineplot(data=df['Tot_Bedrooms'],label="Total Bedrooms")
```

```python
#2
sns.barplot(x=df.Median_Age, y=df.query('Median_Income<6').Median_Income)
```

```python
#3
corr=df.corr()
dataplot=sns.heatmap(df.corr())
```

```python
#4
sns.scatterplot(data=df, x=df['Median_Income'], y=df['Median_House_Value'],hue='Median_Age')

```

```python
#5
rk=df.query('Median_Age<40'and'Median_Age>30' ).Median_Age
sns.kdeplot(data=df,x='Distance_to_LA',hue=rk, shade=True)
```
