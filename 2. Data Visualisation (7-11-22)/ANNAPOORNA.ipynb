```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
from pandas import read_csv
path="/content/California_Houses.csv"
data=read_csv(path)
```

```python
data.head()
```

```python

sns.lineplot(data=data['Tot_Rooms'], label="Total Rooms")
sns.lineplot(data=data['Tot_Bedrooms'], label="Total Bedrooms")
```

```python

plt.title("Median Age v/s Median Income bar graph")

sns.barplot(x=data.Median_Age, y=data.query('Median_Income<6').Median_Income)
```

```python
dataplot=sns.heatmap(data.corr(),annot=True)
plt.show()
```

```python
sns.scatterplot(x=data['Median_Income'], y=data['Median_House_Value'],hue=data['Median_Age'])
```

```python
sns.kdeplot(data=data, x='Distance_to_LA', hue=data.query('30<=Median_Age<=40').Median_Age, shade=True)
```

```python
import plotly.express as px
fig = px.scatter(data, x='Median_House_Value',y='Tot_Rooms', color="Population")
fig.show()
```
