```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("California_Houses.csv")
```

```python
df.head()
```

Qn.1

```python
sns.lineplot(data=df['Tot_Rooms'], label="Total Rooms")

sns.lineplot(data=df['Tot_Bedrooms'], label="Total Bedrooms")
plt.show()
```

Qn.2

```python
sns.barplot(x=df.Median_Age, y=df.query('Median_Income<6').Median_Income)
```

Qn.3

```python
corr=df.corr()
sns.heatmap(corr)
```

Qn.4

```python
sns.scatterplot(x=df['Median_Income'], y=df['Median_House_Value'], hue=df['Median_Age'])
```

Qn.5

```python
sns.kdeplot(x=df['Distance_to_LA'], hue=df.query('Median_Age<=40&Median_Age>=30').Median_Age, fill=True)
```
