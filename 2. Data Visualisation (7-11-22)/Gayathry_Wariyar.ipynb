```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv("/content/California_Houses.csv")
```

```python
df.head()
```

```python
sns.lineplot(data=df['Tot_Rooms'], label="Total Rooms")
sns.lineplot(data=df['Tot_Bedrooms'], label="Total Bedrooms")
```

```python
ans = df.query('Median_Income	< 6').Median_Income
plt.figure(figsize=(10,6))
sns.barplot(x=df['Median_Age'], y=ans)
```

```python
plt.figure(figsize=(14,7))
sns.heatmap(data = df.corr(), annot =True)
```

```python
plt.figure(figsize=(14,7))
sns.scatterplot(x=df['Median_Income'], y=df['Median_House_Value'], hue=df['Median_Age'])
```

```python
plt.figure(figsize=(14,7))
cond_age = df.query('30 <= Median_Age <= 40').Median_Age
sns.kdeplot(data=df, x='Distance_to_LA', hue=cond_age, shade=True)
```

```python
plt.figure(figsize=(14,7))
fig = px.scatter(df, x='Population',y='Distance_to_LA', color="Median_House_Value")
fig.show()
```
