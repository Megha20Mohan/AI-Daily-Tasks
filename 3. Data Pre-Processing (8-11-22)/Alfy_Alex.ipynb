```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("https://raw.githubusercontent.com/ACM-Amrita-Amritapuri/AI-Daily-Tasks/main/Data%20Pre-Processing%20(8-11-22)/Data_Set.csv")
```

```python
data=data.dropna(axis=1)
```

```python
df=pd.DataFrame(data)
```

```python
no_features = len(df.axes[1])
no_features
```

```python
label_encoder = preprocessing.LabelEncoder()
df['ObservationDate']= label_encoder.fit_transform(df['ObservationDate'])
df['Country/Region']= label_encoder.fit_transform(df['Country/Region'])
df['Last Update']= label_encoder.fit_transform(df['Last Update'])
```
