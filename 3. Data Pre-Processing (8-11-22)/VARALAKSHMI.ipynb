```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
df['ObservationDate']= label_encoder.fit_transform(df['ObservationDate'])
df['Country/Region']= label_encoder.fit_transform(df['Country/Region'])
df['Last Update']= label_encoder.fit_transform(df['Last Update'])
df.dtypes
```
